```python
datosMayo = pd.read_csv('2017-05.csv', header = 0)
datosJunio = pd.read_csv('2017-06.csv', header = 0)
datosJulio = pd.read_csv('2017-07.csv', header = 0)
```

```python
def asigvariable (d):
    Edad = d['Edad_Usuario']
    F_R = d['Fecha_Retiro']
    F_A = d['Fecha_Arribo']
    H_R = d['Hora_Retiro']
    H_A = d['Hora_Arribo']
    E_R = d['Ciclo_Estacion_Retiro']
    E_A = d['Ciclo_Estacion_Arribo']
    F_R = d['Fecha_Retiro']
    F_A = d['Fecha_Arribo']
    Genero = d['Genero_Usuario']
    return Edad, F_R, F_A, H_R, H_A, E_R, E_A, F_R, F_A, Genero
    
```

```python
Mayo = asigvariable(datosMayo)
Junio = asigvariable(datosJunio)
Julio = asigvariable(datosJulio)
```

```python
def seleccion_estacion_retiro_hora(estacion, datos):
    ciclo_estaciones = []
    for i in range(len(datos[5])):
        if datos[5][i] == estacion:
            a = [datos[4][i], datos[6][i]]
            ciclo_estaciones.append(a)
    return ciclo_estaciones

def seleccion_estacion_arribo_hora(estacion, datos):
    ciclo_estaciones = []
    for i in range(len(datos[6])):
        if datos[6][i] == estacion:
            a = [datos[3][i], datos[5][i]]
            ciclo_estaciones.append(a)
    return ciclo_estaciones

def seleccion_estacion_retiro_fecha(estacion, datos):
    ciclo_estaciones = []
    for i in range(len(datos[5])):
        if datos[5][i] == estacion:
            a = [datos[7][i], datos[6][i]]
            ciclo_estaciones.append(a)
    return ciclo_estaciones

def seleccion_estacion_arribo_fecha(estacion, datos):
    ciclo_estaciones = []
    for i in range(len(datos[6])):
        if datos[6][i] == estacion:
            a = [datos[8][i], datos[5][i]]
            ciclo_estaciones.append(a)
    return ciclo_estaciones
```

```python
def horas1(a):
    mm = []
    for i in range(len(a)):
        if len(a) == 8:
            b = a[0]
            c = a[1]
            d = a[3]
            e = a[4]
            f = a[6]
            g = a[7]
        if len(a) == 7:
            b = '0'
            c = a[0]
            d = a[2]
            e = a[3]
            f = a[5]
            g = a[6]
    t = b + c + d + e
    tt = int(t)
    mm.append(tt)
    mm.sort()
    return mm

def fecha1(dato, d = 0):
    fe = []
    a = dato[0]
    b = dato[1]
    if d == 1:
        c = dato[4]
        f = c + a + b
    else:
        f = a + b
    ff = int(f)
    fe.append(ff)
    return fe

def Fechas_a_enteros(datos):
    fechas_como_enteros = []
    for i in range(len(datos)):
        fr = fecha1(datos[i][0])
        fa = fecha1(datos[i][1])
        F = fr, fa, datos[i][2] 
        fechas_como_enteros.append(F)
    return fechas_como_enteros

def Dias_a_enteros(datos, d = 0):
    fechas_como_enteros = []
    for i in range(len(datos)):
        if d == 1:
            fr = fecha1(datos[i][0], d =0)
            fr.append(datos[i][1])
        if d == 0:
            fr = fecha1(datos[i][0])
            fr.append(datos[i][1])
        fechas_como_enteros.append(fr)
    return fechas_como_enteros
```

```python
def Clusters(datos, Dimension = 0):

    ms = MeanShift()
    ms.fit(datos)
    etiquetas = ms.labels_
    centros = ms.cluster_centers_

    print (centros)

    n_clusters_ = len(np.unique(etiquetas))

    print ("Número de clusters estimados:", n_clusters_)

    if Dimension == 1:
        colores = 10*('r.','g.','y.','k.','m.','c.','b.')
        
        fig1=plt.figure(figsize=(15,11))
        for i in range(len(datos)):
            plt.plot(datos[i][0], datos[i][1], colores[etiquetas[i]], markersize = 10)
    
        plt.scatter(centros[:,0], centros[:,1], marker = "x", s=150, linewidths= 5, zorder=10)
        #plt.xlim(0,32)
        #plt.ylim(0, 32)
```
